### Implementation

Decision tree classifiers are an algorithm that predicts a target variable based on several input variables. The internal nodes represent input variables that have binary splits depending on a certain condition on that input variable. The split-off node then split off into 2 branches based off of how the information responds to that condition. This repeats in a recursive, cascading effect where you check each internal node's condition until you reach the "leaf" nodes where no more splitting can occur; these leaf nodes are the class labels. 

How the algorithm optimizes for what features and feature thresholds to split the dataset on. How well a split is made is based off of its splitting criteria. There are 2 main criteria for classification: Gini impurity and information gain. We'll discuss this later. 

The main logic behind how to implement decision tree classification is through this recursive process: 
1) Selecting a feature with highest information gain
2) Splitting the dataset based on the values on that feature
3) Repeat this process until either the decision tree can no longer be split further or everything can be adequately classified

#### Gini impurity and information gain criteria

As mentioned before, the splitting criteria are what optimizes the decision tree. Our 2 main methods are the Gini impurity and information gain criteria. Both metrics help us choose the best feature to split on at each node. The feature that gives us the highest information gain (or lowest weighted Gini impurity after the split) becomes our splitting criterion.

Gini impurity measures the probability of incorrectly classifying a randomly chosen element if it was randomly labeled according to the distribution of labels in the subset. It ranges from 0 (perfectly pure) to 0.5 (maximum impurity for binary classification).

$$\text{Gini}(S) = 1 - \sum_{i=1}^{C} p_i^2$$

where $p_i$ is the probability of class $i$ in set $S$, and $C$ is the number of classes.
<br>
<br>
Information gain is based on the concept of entropy from information theory. Entropy measures the average amount of information needed to identify the class of a sample. Higher entropy means more disorder or uncertainty.

$$H(S) = -\sum_{i=1}^{C} p_i \log_2(p_i)$$

Information gain is the reduction in entropy after a split:
$$IG(S, A) = H(S) - \sum_{v \in \text{Values}(A)} \frac{|S_v|}{|S|} H(S_v)$$

where $A$ is the attribute we're splitting on, and $S_v$ is the subset of $S$ where attribute $A$ has value $v$.

```python
# Function that calculates the gini coefficient for a specific array (which would be the node)
def calculate_gini(y):
    # y represents an array of labels at any given node
    
    if len(y) == 0: # Check for pure node and return it if so
        return 0

    n = len(y) # Total elements count
    c = np.unique(y) # Total number of classes/labels
    gini = 0.0


    # Summing logic, add up label proportions for all labels
    for label in c: 
        temp = 0.0
        for i in range(len(y)):
            if y[i] == label:
                temp+= 1

        gini += (temp/n) * (temp/n)


    return 1 - gini # Make proper changes to correctly output gini value

# Function to calculate the entropy, we can calculate information loss by calculating the changes in entropy
    #so we can perform those operations when training our decision tree
def calculate_entropy(y):
    # y represents an array of labels at any given node

    if len(y) == 0:
        return 0

    n = len(y) # Total elements count
    c = np.unique(y) # Total number of classes/labels
    entropy = 0.0


    # Summing logic, add up label proportions for all labels
    for label in c: 
        temp = 0.0
        for i in range(len(y)):
            if y[i] == label:
                temp+= 1

        entropy -= (temp/n) * np.log2(temp/n)


    return entropy # Return the entropy of the array y    
```

#### Handling both categorical and numerical features

The Car Evaluation dataset we will use has features that are all categorical (e.g. "buying price" has vhigh, high, med, low; and "safety" has high, med, low). Decision trees need numerical values to make splits, so we need to convert these text labels into numbers.

I'm encoding the labels, which means I assign each unique category a number. For example:
- "safety" has values [low, med, high], I'll map them to [0, 1, 2]
- "buying" has values [vhigh, high, med, low], I'll map them to [0, 1, 2, 3]

This works well for decision trees because the algorithm doesn't assume any mathematical relationship between the numbers (they don't consider ordinal vs nominal). When the tree splits on "safety <= 1", it's really just separating the data into two groups: {low, med} vs {high}. The tree figures out which groupings are meaningful through the information gain calculations.

One thing to note: this approach wouldn't work for algorithms like linear regression where the numbers would be treated as having magnitude (where 2 is literally "twice" 1). But for decision trees, we're just creating labels that allow the tree to partition the data.

```python
# This has potential problems, should return here before submission

def EncodeCategorical(X, y):
    # X represents the matrix of predictors
    # y represents the target array
    
    # Convert categorical features to numerical
    # Passing by value instead of reference
    X_encoded = X.copy()
    y_encoded = y.copy()
    
    # Dictionary to store the mappings for later reference
    feature_mappings = {}
    
    # We loop over each column 
    for col in X_encoded.columns:

        # Skip columns that are numeric
        if is_numeric_dtype(X_encoded[col]) == True:
            feature_mappings[col] = "numeric"
            continue
        
        # Get all unique values in this column
        unique_vals = X_encoded[col].unique()
        
        # Create a temporary mapping dictionary for current feature
        mapping = {}
        for label in range(len(unique_vals)):
            mapping[unique_vals[label]] = label # This maps all unique values to a number 
        
        # Store the mapping
        feature_mappings[col] = mapping
        
        # Create a temporary array, then append the encoded values in that column 
        encoded_col = []
        for val in X_encoded[col]:
            encoded_col.append(mapping[val])
    
        # Replace the previous column with the encoded column
        X_encoded[col] = encoded_col # Our final matrix of encoded categorical variables as arbitrary numeric type
    
    
    # Encode target variable the same way
    unique_targets = y_encoded['class'].unique()
    target_mapping = {}
    for label in range(len(unique_targets)):
        target_mapping[unique_targets[label]] = label
    
    # Apply target mapping
    encoded_targets = []
    for val in y_encoded['class']:
        encoded_targets.append(target_mapping[val])
    
    y_encoded['class'] = encoded_targets # Final array of the target variables


    return X_encoded, y_encoded, feature_mappings, target_mapping # Return encoded X and y with their mappings
```

#### Recursive tree construction with depth and minimum split size constraints

In order to prevent overfitting, we can place the following 2 constraints: a maximum depth the tree is allowed to extend to (which is a limit on the longest length from root node to leaf node) and a minimum split size (which ensures that a child's node (leaf or otherwise) can have no fewer than a certain threshold of samples. 

This is crucial as while we want to classify the data in much detail, we don't want it to start classifying on noise. By stopping it "early", we allow for the algorithm to narrow down the data while allowing it to remain somewhat generalized in the leaf nodes. 

In order to implement these, we have to include 'maximum depth' and 'minimum split size' as hyperparamters in our Decision Tree algorithm. We will build this into our recursive "Building Tree Function" as we need to keep track of the current depth as well as ensure that an adequate number of samples would be in the current node's children. 

```python
# AI was used in part here for intuition

# Split the dataset based on a feature and threshold
def split_data(X, y, column, threshold):
    # X represents feature matrix
    # y represents targets
    # column represents the feature column we're splitting on 
    # threshold is the value we're splitting on 


    # Compare column values against threshold
    left_indices = X[:, column] <= threshold # Boolean arrays for left branch
    right_indices = X[:, column] > threshold # Boolean arrays for right branch
    
    return left_indices, right_indices    # Returns indices for left and right splits
```

```python
# AI was used in part here for intuition

# Calculates tree split based on information criteria from best feature and threshold to split on
def find_best_split(X, y, criterion):
    # X represents feature matrix
    # y represents targets
    # criterion represents gini or entropy (what information criteria we split on)


    # Initialize local variables that track the best feature and threshold based on best gain
    best_gain = -1
    best_feature = None
    best_threshold = None

    
    n_features = X.shape[1]
    
    # Calculate initial impurity, current_impurity will become a global variable tracking the best 
    if criterion == 'gini':
        current_impurity = calculate_gini(y)
    elif criterion == 'entropy':  # use entropy instead of gini
        current_impurity = calculate_entropy(y)
    else:
        # Throw an error
        raise ValueError("Only support for Gini and Entropy/Information Gain")
    
    # Nested for loop which tests every feature/column and threshold
    for column in range(n_features):
        
        # Get unique values for this feature
        unique_values = np.unique(X[:, column])
        
        # Brute force attempt of threshold on that given feature 
        for threshold in unique_values:
            
            # Split the data
            left_indices, right_indices = split_data(X, y, column, threshold)
            
            # Check if all data would split to left or right branch as there's no point splitting
            if np.sum(left_indices) == 0 or np.sum(right_indices) == 0:
                continue # go to next loop iteration
            
            # Calculate impurity for left and right splits
            y_left = y[left_indices]
            y_right = y[right_indices]
            
            if criterion == 'gini': # gini
                left_impurity = calculate_gini(y_left)
                right_impurity = calculate_gini(y_right)
            else:  # entropy
                left_impurity = calculate_entropy(y_left)
                right_impurity = calculate_entropy(y_right)
                
            
            # Calculate weighted average impurity
            n = len(y)
            n_left = len(y_left)
            n_right = len(y_right)
            weighted_impurity = (n_left / n) * left_impurity + (n_right / n) * right_impurity
            
            # Calculate information gain
            gain = current_impurity - weighted_impurity
            
            # Update best split if current gain is better
            if gain > best_gain:
                best_gain = gain
                best_feature = column
                best_threshold = threshold
    
    return best_feature, best_threshold, best_gain  # Returns the best split 
```

```python
def find_majority_class(y):
    # y represents the array of class labels at a given node 

    # Local variables to track 
    c = np.unique(y) # Array of unique labels
    counts = np.zeros(len(c)) # Track counts for each label, initialized to 0s
    most = -1 # Tracks best count
    best = -1 # Tracks best label

    # Loop through all of y
    for i in range(len(y)): 
        label = y[i]
        for j in range(len(c)): # Loop to identify which label and increment that label's count
            if c[j] == label:
                counts[j] += 1
                break # Applied count, can move on to next label in y
                
    # Find index with highest count
    for i in range(len(counts)):
        if counts[i] > most:
            most = counts[i]
            best = i

    return c[best]  # Return the best label based on its index from counts
```

```python
# AI was used in part for intuition here, specifically with utilizing dictionaries for the nodes

def build_tree(X, y, criterion, max_depth, min_samples_split, current_depth):
    # X represents feature matrix
    # y represents targets
    # criterion represents gini or entropy (what information criteria we split on)
    # max_depth represents the stopping criteria limiting max branch length
    # min_samples_split represents the stopping criteria ensuring that nodes aren't too small or overspecified
    # current_depth represents length of current branch - this helps with keeping track of depth when building recursively
        # specifically for checking with max_depth stopping criterion
    
    # Local variables (to this branch) needed for stopping criteria 
    n_samples = len(y)
    n_classes = len(np.unique(y))
    
    # Stopping criteria, we need to check these first before splitting
        # Stops if max depth reached, not enough samples to split, or pure node
    if (max_depth is not None and current_depth >= max_depth) or n_samples < min_samples_split or n_classes == 1:
        # Reaching here means that stopping criteria has been hit, so spawn leaf node and stop recursion
        
        # Create leaf node with majority class
        leaf_value = find_majority_class(y)     
        return {'type': 'leaf', 'value': leaf_value}

    # Stopping criteria has not been met here, thus we continue splitting
    
    # Find best split
    best_feature, best_threshold, best_gain = find_best_split(X, y, criterion)
    
    # If no good split can be made, but didn't get caught by the earlier stopping criteria, we
        # catch it here and create a leaf node; Claude suggestion
    if best_feature is None or best_gain == 0:
        leaf_value = find_majority_class(y)     
        return {'type': 'leaf', 'value': leaf_value}
    
    # Call our split_data helper function to split
    left_indices, right_indices = split_data(X, y, best_feature, best_threshold)

    
    # Make recursive call to subtrees in the next depth 
    left_subtree = build_tree(X[left_indices], y[left_indices], criterion, 
                             max_depth, min_samples_split, current_depth + 1)
    right_subtree = build_tree(X[right_indices], y[right_indices], criterion, 
                              max_depth, min_samples_split, current_depth + 1)

    
    # Create internal node (Claude)
    return {                         # We use dictionaries as our data structure for our nodes
        'type': 'internal',          # Identifies that the node is not a leaf node
        'feature': best_feature,     # Identifies which feature should be checked at this node
        'threshold': best_threshold, # Identifies what value for that feature to check at this node
        'left': left_subtree,        # Gives left child of this node
        'right': right_subtree       # Gives right child of this node
    }
```

#### Prediction on test samples

Now, we create functions that make predictions based on the nodes features and run it through our decision tree in order to get a predicted value. We apply that to every sample and use that information to determine if it's a good predictor.

We create a predict function which iterates through all the samples and gets their predictions through calling helper function traverse_tree. Our result is an array of label predictions as ints. 

```python
def predict(X, tree):
    # X represents the matrix of predictor values
    # tree represents our decision tree (although it's technically "just" the root node)

    n = X.shape[0] # total sample size
    y_predict = np.zeros(n) # our predictions array
    
    # For loop across the samples 
    for i in range(n):
        y_predict[i] = traverse_tree(tree, X[i]) # We pass our root node and the current sample and store prediction in y_predict

        
    return y_predict.astype(int) # returns explicitly integers for the label predictions
```

```python
# Recursive logic which eventually retrieves the preidction value
def traverse_tree(current_node, sample):
    # current_node represents a dictionary representing the internal node that we're given
        # we're meant to give it the next child or the value 
    
    # Check if node is a leaf and return predicted label, this will "terminate" the recursion
    if current_node['type'] == 'leaf':
        return current_node['value']

    # Get next child based on checking current node's features and values
    feature = current_node['feature']
    threshold = current_node['threshold']

    # Check if the sample's value for specified feature meets threshold
    if(sample[feature] <= threshold):
        return traverse_tree(current_node['left'], sample) # Gives recursive left child subtree
    else:
        return traverse_tree(current_node['right'], sample) # Gives recursive right child subtree 
```

#### Train/Test Split

Next, we will split the data into training and test datasets.

```python
def train_test_split(X, y, train_ratio=0.8, test_ratio=0.2, random_seed = None):
    # X represents our predictor value matrix
    # y represents are an array of target values
    # train_ratio represents the percentage of data that will be used to train our decision tree
    # test_ratio represents the percentage of data that will be used to test our tree
    # random_seed allows for reproducibility, default is no seed


    # Verify that the ratios sum to 1.0, explicit error message and prompt exit if incorrect
    total_ratio = train_ratio + test_ratio
    if abs(total_ratio - 1.0) > 1e-6:
        raise ValueError("Error: Ratios must sum to 1.0")
        
    
    # Set random seed if provided for reproducibility
    if random_seed is not None:
        np.random.seed(random_seed)
    
    n = X.shape[0]  # Total number of samples
    
    # Create array of indices from 0 to n-1 and shuffle them randomly
    indices = np.arange(n)
    np.random.shuffle(indices)  # In-place randomization

    
    # Calculate the split points based on ratios
    train_end = int(n * train_ratio)
    # Remaining samples go to test set
    
    # Split the indices into two groups
    train_indices = indices[:train_end]
    test_indices = indices[train_end:]
    
    # Apply the shuffled indices to X and y to create the splits
    X_train = X[train_indices]
    X_test = X[test_indices]
    
    y_train = y[train_indices]
    y_test = y[test_indices]
    
    return X_train, X_test, y_train, y_test   # Outputs the split datasets according to shuffled indices where X and y rows align
```

#### Evaluate Performance

We will evaluate the performance of our decision trees with the following metrics: Accuracy, precision, recall, F1-Score, and also by creating a Confusion Matrix. 

Accuracy is the measure of the proportion of correct predictions made by the model out of the total number of predictions. It has the following formula: 
<center>$\text{Accuracy} = \frac{\text{Number of Correct Predictions}}{\text{Total Sample Size}}$</center> <br>

Precision is the measure of the instances where a model predicted a value as positive, how many were actually positive. High Precision means that the model doesn't make too many false positives. It has the following formula: 
<center>$\text{Precision} = \frac{TP}{TP + FP}$</center>
<br>
Recall is the measure of the instances where an outcome is actually positive, how often did the model correctly predict it was positive. High Recall means the model doesn't make too many false negatives. It has the following formula:
<center>$\text{Recall} = \frac{TP}{TP + FN}$</center>
<br>
F1-Score is the harmonic mean of precision and recall and is meant to balance both the considerations of the model's recall and its precision. It measures the balance between precision and recall, which makes it a better evaluation tool than accuracy as the effects of false positives and false negatives often differ (i.e. one can have a more severe negative impact over the other). The formula is as follows: 
<center>$\text{F1-Score} = 2 * \frac{Precision * Recall}{Precision + Recall}$</center>
We will calculate the F1-score by using macro-averaged Precision and Recall values. <br>
<br>
The confusion matrix is a table that shows how well our decision tree classifier performed across all classes. For a multi-class problem (like the Car Evaluation dataset), with "n" classes we get an n-by-n matrix. Each row represents the actual class and each column represents the predicted class. The cell at position [i, j] tells us how many samples with actual class i were predicted as class j. <br>
<br>

```python
def compute_accuracy(y_true, y_pred):
    # y_true is the array of the correct values for the targets
    # y_pred is the array of the predicted values for the targets
    
    correct = 0.0 # Tracks number of correct predictions

    # Loop through and sum how many correct predictions
    for i in range(len(y_true)):
        if y_true[i] == y_pred[i]:
            correct += 1

    return correct / len(y_true) # return proportion correct overall total 


def compute_precision(y_true, y_pred):
    # y_true is the array of the correct values for the targets
    # y_pred is the array of the predicted values for the targets

    # local variables to track total precision and # of unique class labels
    unique_class = np.unique(y_true)
    total_precision = 0.0

    # Loop through each unique class label  
    for c in unique_class:
        tp = 0.0
        fp = 0.0
        # loop through all values in predicted y's 
        for i,v in enumerate(y_pred):
            if c == v and v == y_true[i]: # If value is of class 'c' and prediction is correct, it's true positive
                tp += 1
            elif c == v and v!= y_true[i]: # If value is of class 'c' and prediction is incorrect, it's false positive
                fp += 1

        # Ensure that no division by 0 (if no predictions of a certain class in y_pred that exists in y_true)
        if tp + fp > 0:
            total_precision += (tp/(tp + fp))
        else:
            total_precision += 0.0

    return total_precision / len(unique_class) # Return macro-averaged precision value 
    

def compute_recall(y_true, y_pred):
    # y_true is the array of the correct values for the targets
    # y_pred is the array of the predicted values for the targets

    unique_class = np.unique(y_true)
    total_recall = 0.0

    # Loop through each unique class label  
    for c in unique_class:
        tp = 0.0
        fn = 0.0
        # loop through all values in actual y's, so that we get false negative when comparing y_true[i] with y_pred[i]
        for i,v in enumerate(y_true):
            if c == v and v == y_pred[i]: # If value is of class 'c' and prediction is correct, it's true positive
                tp += 1
            elif c == v and v!= y_pred[i]: # If value is of class 'c' and prediction is incorrect, it's false negative
                fn += 1

        # No need for division by 0 check as it'll loop through all values of y_true
        # if tp + fn > 0:
        #     total_recall += (tp/(tp + fn))
        # else:
        #     total_recall += 0.0
        total_recall += (tp/(tp + fn))

    return total_recall / len(unique_class) # Return macro-averaged recall value


def compute_f1score(y_true, y_pred):
    # y_true is the array of the correct values for the targets
    # y_pred is the array of the predicted values for the targets

    # Get Precision and Recall from previous implementations
    precision = compute_precision(y_true, y_pred)
    recall = compute_recall(y_true, y_pred)

    # Check for division by 0 (if it happens to catastrophically predict EVERYTHING wrong, then it'll divide by 0)
    if precision + recall > 0:
        return 2 * ((precision * recall)/(precision + recall))
    else:
        return 0.0


def compute_confusion_matrix(y_true, y_pred):
    # y_true is the array of the correct values for the targets
    # y_pred is the array of the predicted values for the targets

    c = len(np.unique(y_true)) # Number of classes
    
    confusion_matrix = np.zeros((c,c)) # Instantiate n x n multi-class confusion matrix

    # Loop through every sample and see what the actual target is and the predicted target is
    for i in range(len(y_true)):
        actual = y_true[i]
        predicted = y_pred[i]

        confusion_matrix[actual,predicted] += 1 # Increment accordingly with the row number being the actual y 
                                                    # and the column number being the predicted y

    return confusion_matrix
```

### Implementation

We'll start off with implementing a basic logistic regression that uses sigmoid activation

#### Sigmoid Activation

The sigmoid activation function is used to map any real number value into a value in (0,1). We use this as a way to turn raw values into a probability, which we can then utilize in binary classification. For our purposes, we will first create a linear model of our features and then feed that linear combination of features into the sigmoid function. 

The Sigmoid Function is defined as follows: 
<center>$\sigma(z) = \frac{1}{1 + e^{-z}}, \hspace{0.5cm} \sigma(z) \in (0,1)$</center>
where $z$ is a vector of each sample's linear combination of features.<br>

The linear combination $z$ would then be a vector of size n, where n = the number of samples. $z$ is typically expressed as:  
<center>$z = w^{T}x + b, $</center>
where $w$ represents the weights of each feature and $b$ represents the bias. 

The output from $\sigma(z)$ will then be a vector of probabilities, which we then classify into classes "0" and "1" based off a decision boundary: the standard decision boundary is 0.5, so whether $\sigma(z) \geqslant 0.5$.

```python
# Deep-ML was referenced in order to develop intuition 

def sigmoid_logistic_regression(X, weights, bias):
    # X represents the input samples 
    # weights represents the weights 
    # bias represents a bias shift (beta_0)


    # Compute every sample's linear combination using formula Z = Xw + b
    # Then call Sigmoid Function with this array of linear combinations Z
	Z = X @ weights + bias
	probs = sigmoid(Z)

    # Loop through the array of probabilities and classify to Class 1 if sample's probability
    # is greater or equal to 0.5 and Class 0 if less than 0.5
	for i,p in enumerate(probs):
		if(p >= 0.5):
			probs[i] = 1
		else:
			probs[i] = 0

    # Probs is an array with classifications, change type to int
	probs = probs.astype(int)
	return probs    # Output is An array of size N with each sample being classified as a 1 or 0


# Perform sigmoid function calculation
def sigmoid(Z_arr):
    # Z_arr is an array with linear combination for each sample

    # Perform vectorized calculation on our input array
    return 1.0 / (1.0 + np.exp(-Z_arr))  # Outputs an array with the sigmoid function performed (so an array of probabilities)
```

#### Mini-Batch Gradient Descent

There are 3 types of gradient descent that can be used: 
- Batch Gradient Descent: Using all training samples to compute gradients per iteration
- Stochastic Gradient Descent: Uses one sample for gradient computation
- Mini-Batch Gradient Descent: Uses a subset in each computation iteration

For our purposes, we will be using Mini-Batch, because it's not as time or resource intensive as Batch Gradient Descent, but also not as unstable as Stochastic Gradient Descent. 

The loss function for our logistic regression is the log-likelihood, where $p(x_{i})$ is the sigmoid: 
<center>$\sum_{X}y_{i} log(p(x_{i})) + (1 - y_{i}) log[1 - p(x_{i})]$</center><br>

After this, we find the gradient of each feature by taking the partial derivative of the Loss function with respect to each feature weight $\beta_{i}$ and then product by each sample's feature values $x_{i}$ : 
<center>$\nabla Loss = \frac{\partial\mathcal{L}}{\partial\beta_{i}},\hspace{0.5cm} i =$ every feature</center><br>

Finally, taking these gradients, we adjust our $\beta_{i}$'s by the gradient * the learning rate to get the new $\beta_{i}$'s. We repeat this until we have satisfactorily found solid $\beta$'s that minimizes the loss functions.

<br>

So in our implementation, we will perform the following: 
1) Perform an initial parameter guess (we'll start off with 0)
2) Loop and iteratively perform the following on each sample:
   - Compute the predicted probability
   - Compute the total log-likelihood
   - Compute the gradients
   - Update our parameters based on the learning rate and gradient
   - (Eventually) Compute validation loss

```python
# Have to start with arbitrary first guess, initialized all betas to 0
def initialize_parameters(num_parameters):
    # num_parameters represents the number of features/predictors in our data

    # Set all weights and bias to 0 as starting point
    weights = np.zeros(num_parameters)
    bias = 0.0

    return weights, bias   # Outputs the array of 0's for weights and 0.0 for bias terms


# We need to calculate predictions based off of the data, the current weights, and the current bias
def compute_predictions(X, weights, bias):
    # X represents our predictor data
    # weights represents an array of our feature weights (beta_i's)
    # bias represents our bias shift (beta_0)

    # Calculate vectorized predictions by multiplying each sample by given weight and bias 
    y_hat = X @ weights + bias

    # Call previous sigmoid function with current prediction values to turn into a probability (0,1)
    y_hat = sigmoid(y_hat)

    return y_hat   # Outputs an array of probabilities 


# Calculate log-likelihoods to find the loss (then return the average loss)
def compute_LogLikelihood(y_actual, y_predicted):
    # y_actual represents an array of the actual y values (0 or 1) directly from dataset
    # y_predicted represents array of the probabilities in range (0,1) after we computed from sigmoid function
    
    n = y_actual.shape[0] # # of samples
    total_loss = 0.0 # total LL loss

    # Not sure if I can use np.sum or np.mean, so performed manual loop instead of vectorized NumPy Matrix operation, Otherwise: 
    # Perform vectorized calculation of negative log-likelihood (for gradient descent) 
    # loss = -(y_actual * np.log(y_predicted) + (1 - y_actual) * np.log(1 - y_predicted))
    # return np.sum(loss) / n

    # Loop through each value and compute total negative log-likelihood 
    for i in range(n):
        total_loss += -(y_actual[i] * np.log(y_predicted[i]) + (1 - y_actual[i]) * np.log(1 - y_predicted[i]))

    return total_loss / n   # Outputs the average (negative) Log-Likelihood 
    

# Calculate Gradients which are used to update the feature weights and bias
def compute_gradients(X, y_actual, y_predicted):
    # X represents our input predictor data 
    # y_actual represents an array of the actual results from the data (y = 0 or 1)
    # y_predicted represents an array of the predicted results (y-hat in (0,1))

    n = X.shape[0] # sample size of data
    error = y_predicted - y_actual # Array of residuals 

    # Vectorized computation of gradients of all feature weights by taking transpose to take product of errors with 
    #    values from each feature 
    gradients_w = (X.T @ error)/n

    gradient_b = 0
    
    # Bias gradient computed by taking mean bias value 
    # If allowed to use np.sum -> gradient_b = np.sum(error) / n
    for i in range(n):
        gradient_b += error[i]
    gradient_b = gradient_b / n

    
    return gradients_w, gradient_b   # Outputs an array of the weight gradients and the bias gradient


# Update parameters based on gradients
def update_parameters(weights, bias, gradients_w, gradient_b, learning_rate):
    # weights represents array of current feature weights
    # bias represents current bias value
    # gradients_w represents computed gradient for each feature weight
    # gradient_b represents bias gradient
    # learning rate is how big of a step size to change the weight by the gradient

    # Vectorized update of weights by 
    weights = weights - learning_rate * gradients_w
    bias = bias - learning_rate * gradient_b

    return weights, bias   # Outputs updated weights (as an array) and bias value
```

Now that we set the groundwork for our gradient descent, we now implement mini batches. So mini batches works as follows: 
- Take the dataset and randomly shuffle each sample around
- Split the dataset into so-called batches with a 'batch_size' number of samples in each batch
- Loop through a batch individually then update the parameters based only on the gradients calculated from the data in that batch
- Go through the remaining batches, again updating the parameters only by the gradients calculated from that batch alone

```python
# We need to create the mini batches using a function here
def create_minibatches(X, y, batch_size):
    # X represents our predictor data
    # y represents our corresponding target values
    # batch_size represents how large our batches will be

    n = X.shape[0] # number of samples
    mini_batches = [] # Initialized array for our mini batches
    
    # randomize indices for our batches
    indices = list(range(n))
    random.shuffle(indices) # in-place randomization of indices

    # Apply shuffled indices to our X and y
    X_shuffled = X[indices]
    y_shuffled = y[indices]

    # Loop through the samples based on batch size and add batches to the mini_batches array
    for start in range(0, n, batch_size): # This helps us iterate starting values from 0 to n by batch_size "jumps"
     
        end = start + batch_size # end point of each batch
        if(end > n):
            end = n  # Set end point to last element, handles cases where batch_size doesn't divide n equally

        # Apply indices to X and y
        X_batch = X_shuffled[start:end]
        y_batch = y_shuffled[start:end]

        mini_batches.append((X_batch, y_batch)) # Append a tuple of that batch, keeping X and y with same indices aligned


    return mini_batches   # Outputs a list of our mini batches 
```

```python
# ***I had some help from Claude on this code block***

random.seed(55) # Set seed for reproducibility 

# Define a function that can test the functions we have so far on an arbitrary dataset
def train_simple_test(X_train, y_train, learning_rate, batch_size, epochs):
    # X_train represents our predictor data
    # y_train represents our actual classification
    # learning_rate represents our step size we apply with gradient to update our parameters
    # batch_size represents how large our mini batches are 
    # epochs represents the number of iterations across our entire dataset
   
    n_features = X_train.shape[1] # Number of features
    weights, bias = initialize_parameters(n_features) # Call to initialize our parameters

    # The main logic here. We loop 'epochs' number of times, create our mini batches, then compute our predictions
    # (Negative) log-Likelihoods, and our gradients. Then we update our parameters. This represents one 'epoch' iteration
    # Finally we print our loss.
    for epoch in range(epochs):
        
        # Create mini batches every epoch
        batches = create_minibatches(X_train, y_train, batch_size)
        
        # Loop through the mini batches for the given epoch. We make all our remaining function calls here, performed on
        # every mini batch in the current epoch
        for X_batch, y_batch in batches:
            # X_batch and y_batch take on one mini batch tuple of arrays at a time
            
            # Get our y predictors from compute_predictions
            y_predictions = compute_predictions(X_batch, weights, bias)
            
            # Compute gradients 
            grad_w, grad_b = compute_gradients(X_batch, y_batch, y_predictions)
            
            # Update parameters
            weights, bias = update_parameters(weights, bias, grad_w, grad_b, learning_rate)

        
        # Compute loss and print loss 
        y_train_predictions = compute_predictions(X_train, weights, bias)
        print("Epoch: ", epoch + 1, " Loss: ", compute_LogLikelihood(y_train, y_train_predictions))
        

    
    return weights, bias   # Outputs the final weights and bias 


# Dummy data
X_toy = np.array([[1.0, 2.0],
                  [2.0, 3.0],
                  [3.0, 4.0],
                  [4.0, 5.0]])
y_toy = np.array([0, 0, 1, 1])


# Calling our training function to initiate test and get estimated predictor weights and bias
weights, bias = train_simple_test(X_toy, y_toy, 0.1, 2, 20) # Learning rate = 0.1, batch size = 2, 10 epochs

# Check predictions
y_predictions = compute_predictions(X_toy, weights, bias)
print("Predictions: ", y_predictions)
print("Actual:", y_toy)
```

#### L2 Regularization

Regularization, as it pertains to logistic regression, is meant to penalize large weights. Having large weights would result in minute changes causing large changes in our predictions and often means that the data is highly susceptible to noise and overfitting. We will implement L2 regularization, which creates a "penalty" term based off the hyperparameter $\lambda$ multiplied with the squared weights $w_{i}$'s. This $\lambda$ we choose based off of how much penalization we want to perform on for large magnitude weights. 

In order to implement L2 regularization, we would need to modify two functions in our code: compute_gradients and compute_LogLikelihood. 

In compute_gradients, we need to update the gradient formula to be the following: 
<center>$\nabla w = (\frac{1}{n}) * X^{T}$ @ $error + (\frac{\lambda}{n}) * w$</center>
<br>
In compute_LogLikelihood, we need to update the loss function with the additional penalty term applied (partial derivative applied as well): <br>
<br><center>Average Loss = (Total Loss / n) + $(\frac{\lambda}{2n}) * \sum{w^{2}}$</center>

```python
# Calculate log-likelihoods to find the average loss with L2 regularization
def compute_LogLikelihood_withL2(y_actual, y_predicted, weights, lambda_penalty):
    # y_actual represents an array of the actual y values (0 or 1) directly from dataset
    # y_predicted represents array of the probabilities in range (0,1) after we computed from sigmoid function
    # weights represents our feature weights
    # lambda_penalty is our hyperparameter for L2 regularization which affects how severe a penalty to loss will be
    
    n = y_actual.shape[0] # # of samples
    total_loss = 0.0 # total LL loss

    # Not sure if I can use np.sum or np.mean, so performed manual loop instead of vectorized NumPy Matrix operation, Otherwise: 
    # Perform vectorized calculation of negative log-likelihood (for gradient descent) 
    # loss = -(y_actual * np.log(y_predicted) + (1 - y_actual) * np.log(1 - y_predicted))
    # return np.sum(loss) / n

    # Loop through each value and compute total negative log-likelihood 
    for i in range(n):
        total_loss += -(y_actual[i] * np.log(y_predicted[i]) + (1 - y_actual[i]) * np.log(1 - y_predicted[i]))

    # Calculate the L2 penalty
    weights_squared = 0.0
    for weight in weights: 
        weights_squared += weight * weight
    l2_penalty = (lambda_penalty / (2*n)) * weights_squared

    return (total_loss / n) + l2_penalty   # Outputs the average (negative) Log-Likelihood plus the L2 penalty
  

# Calculate Gradients which are used to update the feature weights and bias with L2 regularization
def compute_gradients_withL2(X, y_actual, y_predicted, weights, lambda_penalty):
    # X represents our input predictor data 
    # y_actual represents an array of the actual results from the data (y = 0 or 1)
    # y_predicted represents an array of the predicted results (y-hat in (0,1))
    # weights represents our feature weights
    # lambda_penalty is our hyperparameter for L2 regularization which affects how severe a penalty to the gradients will be

    n = X.shape[0] # sample size of data
    error = y_predicted - y_actual # Array of residuals 

    # Vectorized computation of gradients of all feature weights by taking transpose to take product of errors with 
    #     values from each feature 
    gradients_w = (X.T @ error)/n

    # Calculate our L2 penalty, which is lambda_penalty / n, then apply to weights and add to gradients
    l2_penalty = (lambda_penalty/n) * weights
    gradients_w += l2_penalty
    

    gradient_b = 0
    
    # Bias gradient computed by taking mean bias value (bias is not affected by L2 penalization)
    # If allowed to use np.sum -> gradient_b = np.sum(error) / n
    for i in range(n):
        gradient_b += error[i]
    gradient_b = gradient_b / n

    
    return gradients_w, gradient_b   # Outputs an array of the weight gradients and the bias gradient
```

#### Stopping Early due to Validation Loss

Implementing early stoppage due to declining validation loss is a regularization technique meant to help prevent overfitting by limiting the number of epochs (full iteration across the entire training set) to train so that it doesn't become overly specific to just that data. This occurs since after a certain point, it begins to memorize the training data rather than learning generalizable patterns. 

To combat this, we track the validation loss after each epoch. If the validation doesn't improve for a specified number of epochs (we define this as the hyperparameter "patience"), training is stopped and we revert the weights from the epoch with best/lowest validation loss.<br>
<br>

In order to implement this, we have to keep track of a few new variables:

- User-Defined Hyperparameter 'patience', which is the number of epochs we allow validation loss to stop before giving up 
- The Best validation loss, which tracks the best performing loss up until then
- A Counter for 'patience', which tracks how many epochs have ellapsed without improvement
- The Best Weights and Bias, which track which weights and bias associated with the best validation loss. We use these weights as our final prediction after validation is stopped

```python
# Implementing the training logic for logistic regression, with early stoppages
def train_logistic_regression(X_train, y_train, X_val, y_val, learning_rate, batch_size, max_epochs, lambda_penalty, patience):
    # X_train represents our training predictor data
    # y_train represents our training target values (0 or 1)
    # X_val represents our validation predictor data
    # y_val represents our validation target values (0 or 1)
    # learning_rate is the hyperparameter dictating the step size for gradient descent
    # batch_size is how large our mini-batches are
    # max_epochs represents the maximum number of training iterations we plan on employing (will reach number if no early stoppage)
    # lambda_penalty is our hyperparameter for L2 regularization
    # patience is the hyperparamter of how many epochs we'll wait for validation loss to improve before stopping the loop
    

    
    n_features = X_train.shape[1]  # Number of features
    weights, bias = initialize_parameters(n_features)  # Initialize parameters to 0
    
    # Early stopping variables
    best_val_loss = float('inf')  # Initialize to infinity (loss is guaranteed to improve)
    best_weights = weights.copy()  # Store best weights (.copy() used because otherwise we passing by reference not value) 
    best_bias = bias  # Store the best bias
    patience_counter = 0  # Count epochs without improvement
    

    # Main training loop
    for epoch in range(max_epochs):
        
        # Create mini batches for current epoch (data gets reshuffled each time)
        batches = create_minibatches(X_train, y_train, batch_size)
        
        # Train on each mini batch
        for X_batch, y_batch in batches:
            # X_batch and y_batch represent one mini batch at a time
            
            # Get predictions for current batch
            y_predictions = compute_predictions(X_batch, weights, bias)
            
            # Compute gradients with L2 regularization
            grad_w, grad_b = compute_gradients_withL2(X_batch, y_batch, y_predictions, weights, lambda_penalty)
            
            # Update parameters from calculated gradients
            weights, bias = update_parameters(weights, bias, grad_w, grad_b, learning_rate)
        
        # After current epoch's mini batches trained on all training data, we evaluate on full training and validation sets
        
        # Compute training loss
        y_train_predictions = compute_predictions(X_train, weights, bias)
        train_loss = compute_LogLikelihood_withL2(y_train, y_train_predictions, weights, lambda_penalty)
        
        # Compute validation loss (this is what we use for early stopping)
        y_val_predictions = compute_predictions(X_val, weights, bias)
        val_loss = compute_LogLikelihood_withL2(y_val, y_val_predictions, weights, lambda_penalty)
        
        # After we calculate our training and validation loss, we now have to check if validation loss has improved
        
        # Early stopping logic, compare if validation loss improved
        if val_loss < best_val_loss:
            # Validation loss improved, yay! :)
            best_val_loss = val_loss
            best_weights = weights.copy()  # Save current weights as best, again passing by value not reference
            best_bias = bias  # Save current bias as best
            patience_counter = 0  # Reset patience counter
        else:
            # Validation loss did not improve :(
            patience_counter += 1
            
            # Check if we've run out of patience
            if patience_counter >= patience:
                break  # We've run out of patience and thus break out of the loop prematurely

    
    return best_weights, best_bias   # Return the best weights and bias (not the final ones)
```

The train/validation/test split function works by first shuffling the indices of our dataset randomly to ensure we don't have any ordering bias in our splits. We then calculate the split points based on our desired ratios (default to 70% train, 15% validation, 15% test) and slice our shuffled indices at those points to create three separate index arrays. Finally, we apply these indices to both X and y simultaneously, which ensures that each sample's features and corresponding label remain aligned across all three datasets.

```python
# A function that takes into the feature matrix and target variables, a ratio of the train/val/test
#    (with defaults 70-15-15), and random seed generator 
def train_val_test_split(X, y, train_ratio=0.70, val_ratio=0.15, test_ratio=0.15, random_seed=None):
    # X represents the feature matrix size number of samples by number of features
    # y represents the target vector with length equal to number of samples
    # train_ratio represents the proportion of data allocated to training set
    # val_ratio represents the proportion of data allocated to validation set
    # test_ratio represents the proportion of data allocated to test set
    # random_seed is an optional parameter for reproducibility

    # Verify that the ratios sum to 1.0, explicit error message and prompt exit if incorrect
    total_ratio = train_ratio + val_ratio + test_ratio
    if abs(total_ratio - 1.0) > 1e-6:
        raise ValueError("Error: Ratios must sum to 1.0")
    
    # Set random seed if provided for reproducibility
    if random_seed is not None:
        np.random.seed(random_seed)
    
    n = X.shape[0]  # Total number of samples
    
    # Create array of indices from 0 to n-1 and shuffle them randomly
    indices = np.arange(n)
    np.random.shuffle(indices)  # In-place randomization

    
    # Calculate the split points based on ratios
    train_end = int(n * train_ratio)
    val_end = train_end + int(n * val_ratio)
    # Remaining samples go to test set
    
    # Split the indices into three groups
    train_indices = indices[:train_end]
    val_indices = indices[train_end:val_end]
    test_indices = indices[val_end:]
    
    # Apply the shuffled indices to X and y to create the splits
    X_train = X[train_indices]
    X_val = X[val_indices]
    X_test = X[test_indices]
    
    y_train = y[train_indices]
    y_val = y[val_indices]
    y_test = y[test_indices]
    
    return X_train, X_val, X_test, y_train, y_val, y_test   # Outputs the split datasets according to shuffled indices where X and y rows align
```

#### Normalize Features

Before training our logistic regression model, we need to normalize our features to ensure that gradient descent converges efficiently. Features in the breast cancer dataset have different scales (for example, area ranges from ~100 to ~2000 while smoothness ranges from ~0.05 to ~0.16), and these different scales can cause issues during training. When features have vastly different magnitudes, gradient descent can oscillate or converge very slowly because the gradients for large-scale features dominate the updates. Normalization addresses this by transforming all features to have mean 0 and standard deviation 1 using the formula: $x_{normalized} = \frac{x - \mu}{\sigma}$, where $\mu$ is the mean and $\sigma$ is the standard deviation.

A critical consideration when normalizing is preventing data leakage. Data leakage occurs when information from the validation or test sets influences the training process, which artificially inflates performance metrics and doesn't reflect real-world model performance. To prevent this, we compute the normalization parameters (mean and standard deviation) using only the training data, then apply those same parameters to normalize the validation and test sets. This ensures that our model has no knowledge of the validation or test distributions during training, which mirrors real-world deployment where we would normalize new incoming data using statistics saved from our original training set.

```python
# Create a function that normalizes all the predictor matrices from the different splits
def normalize_breast_cancer_data(X_train, X_val, X_test):
    # X_train represents training features
    # X_val represents validation features  
    # X_test represents test features
    
    n_train = X_train.shape[0]
    n_features = X_train.shape[1]
    
    # Calculate means from training data only
    means = np.zeros(n_features)
    for j in range(n_features):
        feature_sum = 0.0
        for i in range(n_train):
            feature_sum += X_train[i, j]
        means[j] = feature_sum / n_train
    
    # Calculate standard deviations from training data only
    stds = np.zeros(n_features)
    for j in range(n_features):
        squared_diff_sum = 0.0
        for i in range(n_train):
            diff = X_train[i, j] - means[j]
            squared_diff_sum += diff * diff
        variance = squared_diff_sum / n_train
        stds[j] = np.sqrt(variance)
    
    # Normalize training set
    X_train_norm = np.zeros_like(X_train)
    for j in range(n_features):
        for i in range(n_train):
            if stds[j] > 1e-8:
                X_train_norm[i, j] = (X_train[i, j] - means[j]) / stds[j]
            else:
                X_train_norm[i, j] = 0.0
    
    # Normalize validation set using training statistics
    n_val = X_val.shape[0]
    X_val_norm = np.zeros_like(X_val)
    for j in range(n_features):
        for i in range(n_val):
            if stds[j] > 1e-8:
                X_val_norm[i, j] = (X_val[i, j] - means[j]) / stds[j]
            else:
                X_val_norm[i, j] = 0.0
    
    # Normalize test set using training statistics
    n_test = X_test.shape[0]
    X_test_norm = np.zeros_like(X_test)
    for j in range(n_features):
        for i in range(n_test):
            if stds[j] > 1e-8:
                X_test_norm[i, j] = (X_test[i, j] - means[j]) / stds[j]
            else:
                X_test_norm[i, j] = 0.0
    
    return X_train_norm, X_val_norm, X_test_norm   # Outputs all three normalized datasets
```

#### Perform Evaluation Metrics 

After training our logistic regression model, we need to evaluate its performance on the test set to understand how well it generalizes to unseen data. We will compute five key metrics that each provide different insights into model performance:

Accuracy measures the overall correctness of our model and is defined as the proportion of correct predictions out of all predictions: 
<center>$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$</center>
While accuracy is intuitive, it can be misleading for imbalanced datasets where one class dominates.
<br><br>
Precision measures how many of our positive predictions were actually correct and is defined as: 
<center>$\text{Precision} = \frac{TP}{TP + FP}$</center>
In our context, precision tells us: of all the cases we predicted as malignant, what proportion were actually malignant. High precision means fewer false alarms.
<br><br>

Recall (also called sensitivity or true positive rate) measures how many of the actual positive cases we correctly identified and is defined as:
<center>$\text{Recall} = \frac{TP}{TP + FN}$</center> 
In our context, recall tells us: of all the actual malignant cases, what proportion did we catch. High recall means we're missing fewer malignant cases, which is critical in medical diagnosis.
<br>
<br>
F1-Score is the harmonic mean of precision and recall, providing a single metric that balances both concerns: 
<center>$\text{F1} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}$</center>
The F1-score is particularly useful when you need to balance false positives and false negatives.
<br>
<br>
ROC-AUC (Area Under the Receiver Operating Characteristic Curve) measures the model's ability to distinguish between classes across all possible classification thresholds. The ROC curve plots the true positive rate against the false positive rate at various threshold settings, and the AUC represents the area under this curve. An AUC of 1.0 indicates perfect classification, while 0.5 indicates random guessing. ROC-AUC is especially useful because it's threshold-independent and works well even with imbalanced datasets.

```python
def compute_accuracy(y_true, y_pred):
    # y_true represents the actual labels
    # y_pred represents the predicted labels
    
    correct = 0  # Counter for correct predictions
    total = len(y_true)  # Total number of samples
    
    # Count how many predictions match the actual labels
    for i in range(total):
        if y_true[i] == y_pred[i]:
            correct += 1
    
    accuracy = correct / total
    
    return accuracy   # Outputs accuracy as a float between 0 and 1


def compute_precision(y_true, y_pred):
    # y_true represents the actual labels
    # y_pred represents the predicted labels
    
    true_positives = 0
    false_positives = 0
    
    # Count true positives and false positives
    for i in range(len(y_true)):
        if y_pred[i] == 1:  # Model predicted positive
            if y_true[i] == 1:
                true_positives += 1
            else:
                false_positives += 1
    
    # Avoid division by zero
    if true_positives + false_positives == 0:
        return 0.0
    
    precision = true_positives / (true_positives + false_positives)
    
    return precision   # Outputs precision as a float between 0 and 1


def compute_recall(y_true, y_pred):
    # y_true represents the actual labels
    # y_pred represents the predicted labels
    
    true_positives = 0
    false_negatives = 0
    
    # Count true positives and false negatives
    for i in range(len(y_true)):
        if y_true[i] == 1:  # Actual label is positive
            if y_pred[i] == 1:
                true_positives += 1
            else:
                false_negatives += 1
    
    # Avoid division by zero
    if true_positives + false_negatives == 0:
        return 0.0
    
    recall = true_positives / (true_positives + false_negatives)
    
    return recall   # Outputs recall as a float between 0 and 1


def compute_f1_score(y_true, y_pred):
    # y_true represents the actual labels
    # y_pred represents the predicted labels
    
    # Calculate precision and recall using our previously defined functions
    precision = compute_precision(y_true, y_pred)
    recall = compute_recall(y_true, y_pred)
    
    # Avoid division by zero
    if precision + recall == 0:
        return 0.0
    
    f1 = 2 * (precision * recall) / (precision + recall)
    
    return f1   # Outputs F1 score as a float between 0 and 1


def compute_roc_auc(y_true, y_scores):
    # y_true represents the actual labels (0 or 1)
    # y_scores represents the predicted probabilities from sigmoid function
    
    # Get unique threshold values from our scores
    thresholds = []
    for score in y_scores:
        if score not in thresholds:
            thresholds.append(score)
    
    # Sort thresholds in descending order
    thresholds.sort(reverse=True)
    
    # Initialize lists to store TPR and FPR values
    fpr_list = []
    tpr_list = []
    
    # For each threshold, compute TPR and FPR
    for threshold in thresholds:
        # Make predictions based on current threshold
        y_pred = np.zeros(len(y_scores))
        for i in range(len(y_scores)):
            if y_scores[i] >= threshold:
                y_pred[i] = 1
        
        # Count TP, FP, TN, FN for this threshold
        tp = 0
        fp = 0
        tn = 0
        fn = 0
        
        for i in range(len(y_true)):
            if y_true[i] == 1 and y_pred[i] == 1:
                tp += 1
            elif y_true[i] == 0 and y_pred[i] == 1:
                fp += 1
            elif y_true[i] == 0 and y_pred[i] == 0:
                tn += 1
            elif y_true[i] == 1 and y_pred[i] == 0:
                fn += 1
        
        # Calculate TPR and FPR
        tpr = tp / (tp + fn) if (tp + fn) > 0 else 0
        fpr = fp / (fp + tn) if (fp + tn) > 0 else 0
        
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    
    # Sort by FPR to ensure proper ordering for AUC calculation
    sorted_indices = sorted(range(len(fpr_list)), key=lambda i: fpr_list[i])
    fpr_sorted = [fpr_list[i] for i in sorted_indices]
    tpr_sorted = [tpr_list[i] for i in sorted_indices]
    
    # Compute area under curve using trapezoidal rule
    auc = 0.0
    for i in range(1, len(fpr_sorted)):
        width = fpr_sorted[i] - fpr_sorted[i-1]  # Width of trapezoid
        height = (tpr_sorted[i] + tpr_sorted[i-1]) / 2.0  # Average height
        auc += width * height  # Add area of trapezoid
    
    return auc   # Outputs AUC as a float between 0 and 1
```

### Implementation 

#### Theory behind SVM, Primal/Dual Formulation, and Kernel Trick

The main objective in SVM is finding the "best" hyperplane that seperates two classes of data. What this entails is maximizing the margin (distance) between classes. SVM can be used for both classification and regression, however, it's typically used for binary classification between two classes.

The best hyperplane is called the decision boundary, and in linear classification, is marked by the following formula:
<center>$w^T \cdot x + b = 0$</center>
where $w$ is the normal vector that defines the hyperplane, $x$ is the set of data points, and $b$ is the bias that represents the hyperplane's shift along the normal vector. <br>
<br>
When we take the dot product of the hyperplane's normal vector with each point (i.e. $w^T \cdot x$), we are taking the projection of each point in the vector of points x with each component in the vector w. Functionally, this looks like <center>$proj_{w}x = w_{1}x_{1} + w_{2}x_{2} + ... + w_{n}x_{n}$, for n components. </center><br>

The margin is then marked by 2 other hyperplanes, defined by support vectors that lie on said hyperplanes, which run parallel to the decision boundary. This total margin is equal to the distance between these 2 hyperplanes. Formally, since bias represents the shift along the normal vector, the distance from any point to the decision boundary is $d = \frac{|w^T \cdot x + b|}{||w||}$.
<br>
For our support vectors that lie on the 2 hyperplanes, we define them to satisfy the following equation: 
<center>$w^T \cdot x + b = \pm1$</center>
simply because while we could have chosen any other constants, this makes the math the cleanest. 
<br>
This then makes the distance between decision boundary and positive support vectors ($d_{+}$) to be $d_{+} = \frac{|+1|}{||w||}$, the distance to the negative support vectors to be $d_{-} = \frac{|-1|}{||w||}$, and the total margin to be: <center>$\text{total margin} = \frac{2}{||w||}$</center><br>
This total margin is what we need to maximize, thus we need to minimize ||$w$|| in order to get the best Support Vectors and Decision Boundary.<br>
<br>
<center>*************************************</center>
<br>
This Max Margin is our cost function and is subject to the following constraints: 
<center>for $y_i = +1$, $w \cdot x_i + b \geq 1$ </center>
<center>for $y_i = -1$, $w \cdot x_i + b \leq -1$ </center><br>
where $Max(w,b)$ gets us our largest margin. We can simplify these inequality constraints to be: 
<center>$y_i \cdot (w \cdot x_i + b) \geq 1$, where $y_i = \pm1$  </center><br>

Therefore we must find:
<center>$Max(w,b) \frac{2}{||w||} \hspace{0.5cm} s.t. \hspace{0.25cm} \forall i \hspace{0.5cm} y_i \cdot (w \cdot x_i + b) \geq 1$</center><br>

This is equivalent to: 
<center>$Min(w,b) \frac{||w||}{2} \hspace{0.5cm} s.t. \hspace{0.25cm} \forall i \hspace{0.5cm} y_i \cdot (w \cdot x_i + b) \geq 1$</center><br>

Finallly, because $f: x \mapsto x^2$ is monotonic on $\mathbb{R}^{+}$:
<center>$Min(w,b) \frac{||w||^2}{2} \hspace{0.5cm} s.t. \hspace{0.25cm} \forall i \hspace{0.5cm} y_i \cdot (w \cdot x_i + b) \geq 1$</center><br>
This is our primal formulation, where we're directly optimizing for $w$ and $b$ in the original feature space.<br>
<br><center>*************************************</center>
<br>
We now have $\frac{||w||^2}{2}$ as our objective/loss function, which measures and captures the margin. This is a quadratic term. To solve this constrained optimization problem, we use duality. The dual formulation will allow us to work with dot products only, which we employ later when we implement using kernels. 

We start by constructing a Lagrange equation, adding one Lagrange multiplier $\alpha_i$ for constraint in our original problem:

<center>$L(w, b; \alpha_1, \alpha_2, ..., \alpha_n) = \frac{||w||^2}{2} - \sum_{i=1}^{n} \alpha_i [y_i(wx_i + b) - 1]$</center><br>

Now we apply the first order conditions of optimality (KKT conditions). Taking the partial derivative with respect to $w$ and setting it equal to zero gives us:

<center>$\frac{\partial L}{\partial w} = 0 \implies w = \sum_{i=1}^{n} \alpha_i y_i x_i$</center><br>

This tells us that the optimal $w$ can be expressed as a weighted combination of training points! The weights are now $a_i y_i$ and only support vectors (where $a_i > 0$) contribute. Taking the partial derivative with respect to $b$ and setting it equal to zero gives us:

<center>$\frac{\partial L}{\partial b} = 0 \implies \sum_{i=1}^{n} \alpha_i y_i = 0$</center><br>

Additionally, the complimentary slackness condition from KKT give us that $ \forall i$:

<center>$\alpha_i [y_i(wx_i + b) - 1] = 0$</center><br>

This means either $\alpha_i = 0$ or $wx_i + b = 1$ for each data point. In essence, $wx_i + b = 1$ means that that point $x_i$ is a margin support vector and the points that satisfy $\alpha_i = 0$ are far away from the boundary.<br>
<center>*************************************</center>
<br>

Through the dual formulation, we use the conditions we just derived to eliminate our dependency on $w$ and $b$, leaving us with an optimization problem only in terms of the $\alpha_i$ values.

The idea is to substitute $w = \sum_{i=1}^{n} \alpha_i y_i x_i$ back into our Lagrangian $L(w, b; \alpha_i)$. Starting with our original Lagrangian and expanding it out:

<center>$L(\vec{\alpha}) = \frac{||w||^2}{2} - \sum_{i=1}^{n} \alpha_i [y_i(wx_i + b) - 1]$</center><br>

After substituting and expanding (multiplying out the terms):

<center>$= \frac{1}{2}\left(\sum_{i=1}^{n} \alpha_i y_i x_i\right)\left(\sum_{j=1}^{n} \alpha_j y_j x_j\right) - \sum_{i=1}^{n} \alpha_i \left[y_i\left(\sum_{j=1}^{n} \alpha_j y_j x_j\right) \cdot x_i + b\right] + \sum_{i=1}^{n}\alpha_i$</center><br>

Continuing to simplify and combining like terms:

<center>$= \frac{1}{2}\sum_{i=1}^{n}\sum_{j=1}^{n} \alpha_i \alpha_j y_i y_j (x_i x_j) - \sum_{i=1}^{n}\sum_{j=1}^{n} \alpha_i \alpha_j y_i y_j (x_i x_j) - \sum_{i=1}^{n} \alpha_i y_i b + \sum_{i=1}^{n} \alpha_i$</center><br>

Recall from our KKT conditions that we have the constraint $\sum_{i=1}^{n} \alpha_i y_i b = 0$. This allows us to simplify our expression to:

<center>$L(\vec{\alpha}) = -\frac{1}{2}\sum_{i=1}^{n}\sum_{j=1}^{n} \alpha_i \alpha_j y_i y_j (x_i x_j) + \sum_{i=1}^{n} \alpha_i$</center><br>

The formula here **only depends on the dot products** $x_i x_j$, and we *never* need the explicit weight vector. This is the foundation of the kernel trick. So our dual problem becomes minimizing this expression subject to the KKT conditions we derived earlier:

<center>$\min_{\alpha_i} L(\vec{\alpha}) = -\frac{1}{2}\sum_{i=1}^{n}\sum_{j=1}^{n} \alpha_i \alpha_j y_i y_j (x_i x_j) + \sum_{i=1}^{n} \alpha_i$</center><br>

<center>subject to: $\alpha_i \geq 0$ and $\sum_{i=1}^{n} \alpha_i y_i = 0$</center><br>
This minimization form is what we'll use Scipy for later on.<br>
<br>
<center>*************************************</center>
<br>
To find the optimal solution, we take the partial derivative of $L(\vec{\alpha})$ with respect to each $\alpha_i$ and set it equal to zero. This allows us to find the optimal $\alpha_i^*$ values for all $i$. 

Once we have solved for the optimal $\alpha_i^*$ values, we can plug them back into our expression for $w$ to get:

<center>$w^* = \sum_{i=1}^{n} \alpha_i^* y_i x_i$</center><br>

However, we still need to find $b^*$. To do this, we use the constraint that support vectors satisfy $y_i(w^{*} \cdot x_i + b^{*}) = 1$. Using a support vector $x_s$ with label $y_s$ from our training set, we can solve for: 

<center>$b^{*} = y_s - w^{*T} x_s = y_s - \sum_{j=1}^{n} \alpha_j^* y_j (x_j \cdot x_s)$</center>
<br>

For numerical stability, we typically average over all margin support vectors (those support vecotrs satisfying $ 0 < alpha_i < C$). 

Phew, we finally made it! The pair $(w^*, b^*)$ defines the hyperplanes optimal hyperplanethat that maximizes the margin between our two classes. But wait - this only works for linearly separable data. What if our data needs a non-linear boundary? That's where kernels come in!
<br><center>*************************************</center>

Everything we've done so far assumes our data is linearly separable, meaning that we can draw a straight line (or hyperplane) to separate the classes. But many real-world datasets aren't linearly separable in their original space. 

One solution is to map our data to a higher-dimensional space where it becomes linearly separable. For 2D data $(x_1,x_2)$, we could transform it to:

<center>$\phi(x) = [x_1, x_2, x_1^2, x_2^2, x_1x_2, ...]$</center>
<br>
In this higher-dimensional space, the data might be linearly separable. However, computing $\phi(x)$ explicitly becomes computationally prohibitive or even impossible as the dimensionality grows. For example, Radial Basis Function (RBF) mapping would require infinite dimensions.

That's where we introduce the clever *kernel trick*. The key insight from our dual formulation is that we only need dot products $x_i^T \cdot x_j$. We never need the explicit weight vector $w$ or the mapped features $\phi(x)$.

A kernel function $K(x_i,x_j)$ computes the dot product in the feature space without ever computing the feature mapping:

<center>$K(x_i, x_j) = \phi(x_i)^T \phi(x_j)$</center>
<br>
This means we can compute $K(x_i,x_j)$ directly in the original space, avoiding the expensive or impossible computation of $\phi(x)$. As such, we can simply replace every dot product in our dual formulation with a kernel:

<center>$L(\vec{\alpha}) = \sum_{i=1}^{n} \alpha_i - \frac{1}{2}\sum_{i=1}^{n}\sum_{j=1}^{n} \alpha_i \alpha_j y_i y_j \hspace{0.15cm} K(x_i, x_j)$</center>
<br>
<center>$f(x) = \text{sign}\left(\sum_{i=1}^{n} \alpha_i y_i K(x_i, x) + b\right)$</center>
<br>
Notice that $w$ never appears explicitly. The decision boundary is defined implicitly through support vectors and the kernel function.<br>
<br>
<center>*************************************</center>
<br>

For linear kernels, it's simply just the regular dot product. No transformation is needed, but the data does have to be linearly separable for it to properly function. 

<center>$K(x_i, x_j) = x_i^T x_j$</center>
<br>

For non-linear kernels, that's a completely different story. For our implementation here, we will employ RBF, which maps our original space to infinite-dimensional feature space. Our RBF kernels look like the following:
<center>$K(x_i, x_j) = \phi(x_i)^T \phi(x_j)$</center>
<br>
and despite the fact that it's in infinite dimensions, we can utilize the kernel trick to compute it directly by using the following expression, which calculates the Euclidean distance between the two points in the original space: 
<br>
<center>$\text{exp}(\gamma||x_i - x_j||^2)$</center>
<br>
When $x_i$ and $x_j$ are close: $||x_i - x_j||^2 \approx 0 \implies K(x_i,x_j) \approx 1$, or that they have high similarity<br>
When $x_i$ and $x_j$ are far apart: $||x_i - x_j||^2 \text{is large} \implies K(x_i,x_j) \approx 0$, or that they have low similarity<br>
<br>
Each support vector creates a Gaussian "bump" centered at that point. The hyperparameter $\gamma$ controls the width of these bumps: A large $\gamma$ creates narrow bumps which risk overfitting and a small $\gamma$ creates wide bumps which risk underfitting. 

#### Kernel Function

```python
# Calculate the scalar based off the kernel type and 2 vectors x_i and x_j
def kernel(x_i, x_j, kernel_type, gamma=0.1):
    # x_i is the first vector of points
    # x_j is the second vector of points
    # kernel_type is either "linear" or "RBF", denoting what type of kernel operation to perform
        # on vectors x_i and x_j
    # gamma is the hyperparameter for RBF kernels that controls width
    
    if kernel_type == "linear":
        # Simple Linear Kernels, dot product of the 2 vectors
        return np.dot(x_i, x_j)
        
    elif kernel_type == "rbf":
        # RBF kernel: exp(-gamma * ||x_i - x_j||^2)

        diff = x_i - x_j 

        squared_distance = np.dot(diff, diff)

        return np.exp(-gamma * squared_distance)
        
    else:
        raise ValueError("kernel type currently not supported")
    # Kernel function returns a kernel value, a scalar which describes how similar values are
```

#### Objective Function

```python
# calculate L(alpha) given vectors and matrices
def objective_function(alpha, X, y, kernel_type, gamma=0.1):
    # alpha denotes a vector of the alpha_i's or the lagrangian multipliers
    # X represents the matrix of all n samples by m features
    # y represents the vector of target classifications
    # kernel_type denotes what type of kernel operation to perform
    # gamma is the hyperparameter for RBF kernels that controls width
    
    lagrangian = 0.0

    # first part of objective function, sum of all lagrangian multipliers
    for i in range(len(alpha)):
        lagrangian += alpha[i]

    # second part of objective function, taking the product of a_i, a_j, y_i, y_j, with the kernel
        # of x_i, x_j
    for i in range(len(alpha)):
        for j in range(len(alpha)):
            lagrangian -= 0.5 * (alpha[i] * alpha[j] * y[i] * y[j]) * kernel(X[i], X[j], kernel_type, gamma)
    
    return -lagrangian # return the output of our Lagrangian equation given the chosen alphas, negative to support
                            # using scipy to minimize
```

#### Fitting the Model

```python
# Note, that since we have to use scipy.optimize.minimize(), we will import it here: 
# !pip install scipy
from scipy.optimize import minimize
```

```python
# fit the model based off of input values, the kernel type, and a regularization hyperparameter
    # employs all helpers as well as Scipy's minimize() function to generate alpha*'s used to eventual
    # obtain all support vectors and thus the decision boundary and hyperplanes
def fit(X, y, kernel_type, C=1.0, gamma=0.1):
    # X represents the matrix of all n samples by m features
    # y represents the vector of target classifications
    # kernel_type denotes what type of kernel operation to perform
    # C represents the regularization parameter
    # gamma is the hyperparameter for RBF kernels that controls width

    
    n = X.shape[0] # sample size 

    # Initialize alphas
    alpha = np.zeros(n)
    
    # Establish constraints
    # Constraint: sum of a_i * y_i = 0
    sum_constraint = {'type': 'eq',
                      'fun': lambda alpha: np.dot(alpha, y)  # Claude helped with lambda functions
    }
    
    # Using Scipy.optimize.minimize() to find most optimal set of alpha-star's
        # We pass in our objective function, the initial alphas, arguments for the objective function
        # And the 2 constraints generated from KKT
    result = minimize(
        fun = objective_function, 
        x0 = alpha,
        args = (X, y, kernel_type, gamma), # Arguments passed to our objective_function; Claude helped
        constraints = [sum_constraint],
        method = 'SLSQP', # Ensures optimized for equality constraints and bounds; Claude pointed out this logic
        bounds = [(0, C) for _ in range(n)] # This establishes the condition alpha_i >= 0 for all i
    )

    alpha = result.x # Retrieve vector of optimal alphas

    # Retrieve indices of all support vectors
    # Support vectors are where alpha > 0
    indices = np.where(alpha > 1e-5)[0]

    # Call calculate_bias helper function for bias
    bias = calculate_bias(alpha, X, y, indices, kernel_type, C, gamma)


    # Store model as a dictionary
    model = {
        'alpha': alpha[indices],
        'sv': X[indices],
        'sv_labels': y[indices],
        'bias': bias,
        'kernel_type': kernel_type,
        'C': C, 
        'gamma': gamma
    }
    
    return model # Return our dictionary model that contains all the support vectors and biases   
```

#### Optimizing the Code

So originally in my code, I had my objective function call the kernel function directly in the nested for loop with $x_i$, $x_j$, which would compute $K(X[i],X[j])$ for every (i,j) pair. This is already computationally heavy, being $O(n^2)$. HOWEVER, because the objective function itself is also being called anywhere from 100-500 times when Scipy's minimize was optimizing for $a^*$'s, this would result in $100 * (568 * 0.7)^2$ kernel operations for sample size n = 568!!! That's **~16 MILLION** kernel computations *every* time you call the fit() function.<br>
<br>
Now, optionally I could manually reduce the total sample size in order to reduce the number of computations, but frankly that seemed extremely inelegant and I believe that having my algorithm be $O(n^3)$ was problematic as is, so there was a better solution.<br>
<br>
By precomputing the kernel matrix once *before* optimization, we reduce the number of kernel operations by a factor of $N$, making it $O(N^2)$, or **160,000** kernel computations done once. <br>
<br>
This, however, was still relatively slow :( so the code needed to be optimized further by vectorizing the kernel matrix and objective function calculations.<br>

For calculating the Kernel Matrix, instead of using nested for loops to compute each kernel value individually, we can leverage NumPy's broadcasting capabilities. For linear kernels, this becomes a simple matrix multiplication $X \text{@} X^T$. For RBF kernels, we compute all pairwise squared distances at once using broadcasting: $||X[i] - X[j]||^2$ for all (i,j) pairs can be computed as $||X[i]||^2 + ||X[j]||^2 - 2*X[i] \cdot X[j]$, which eliminates the nested loops entirely. (Claude)

```python
# # Function to precompute kernel matrix
# def compute_kernel_matrix(X, kernel_type, gamma=0.1):
#     # X represents the training data matrix
#     # kernel_type is kernel operation to be performed
#     # gamma is the hyperparameter for RBF kernels
    
#     n = X.shape[0] # sample size
    
#     # Initialize kernel matrix
#     K = np.zeros((n, n))
    
#     # Compute kernel for all pairs of training points
#     for i in range(n):
#         for j in range(n):
#             K[i, j] = kernel(X[i], X[j], kernel_type, gamma)
    
#     return K # return a matrix of all 400-by-400 kernel computations performed


# Function to precompute kernel matrix using vectorized operations
def compute_kernel_matrix(X, kernel_type, gamma=0.1):
    # X represents the training data matrix
    # kernel_type is kernel operation to be performed
    # gamma is the hyperparameter for RBF kernels
    
    n = X.shape[0] # sample size
    
    if kernel_type == "linear":
        # Perform vectorized matrix multiplication X @ X^T
        K = np.dot(X, X.T)

    ##### I really needed a lot of help with this vectorization process, feel free to penalize for Claude use here#####
    elif kernel_type == "rbf":
        # For RBF kernel: K[i,j] = exp(-gamma * ||X[i] - X[j]||^2)
        # We use the expansion: ||X[i]||^2 + ||X[j]||^2 - 2*X[i]*X[j]
        
        # Compute squared norms for each sample: ||X[i]||^2 for all i
        # This gives us a column vector of shape (n, 1)
        sq_norms = np.sum(X ** 2, axis=1).reshape(-1, 1)
        
        # Using broadcasting to compute all pairwise squared distances at once:
            # sq_norms broadcasts to (n, n) vertically, sq_norms.T broadcasts horizontally
            # X @ X.T computes all dot products X[i]X[j] simultaneously
            sq_dists = sq_norms + sq_norms.T - 2 * np.dot(X, X.T)
        
        # Apply RBF transformation element-wise to entire matrix
        K = np.exp(-gamma * sq_dists)
        
    else:
        raise ValueError("kernel type currently not supported")
        
    return K # return kernel matrix computed with vectorized operations
```

We do need to change our objective and fit functions now in order to accomodate for this. Including getting rid of the kernel calls and changing arguments to accept the *K* kernel matrix rather than "kernel computation" specific arguments. 

```python
# # Optimized objective function that uses precomputed kernel matrix
# def objective_function_optimized(alpha, K, y):
#     # alpha denotes a vector of the alpha_i's or the lagrangian multipliers
#     # K represents the precomputed kernel matrix
#     # y represents the vector of target classifications
    
#     lagrangian = 0.0

#     # First part of objective function, sum of all lagrangian multipliers
#     for i in range(len(alpha)):
#         lagrangian += alpha[i]

#     # Second part of objective function, using precomputed kernel values
#     for i in range(len(alpha)):
#         for j in range(len(alpha)):
#             lagrangian -= 0.5 * (alpha[i] * alpha[j] * y[i] * y[j]) * K[i, j] # Main change is here! Simply access element K[i,j] in O(1) time!
    
#     return -lagrangian # return negative of sum for minimization



# Optimized objective function that uses precomputed kernel matrix
def objective_function_optimized(alpha, K, y):
    # alpha denotes a vector of the alpha_i's or the lagrangian multipliers
    # K represents the precomputed kernel matrix
    # y represents the vector of target classifications
    

    # First part of objective function, sum of all lagrangian multipliers
    first_term = 0.0
    for i in range(len(alpha)):
        first_term += alpha[i]
    
    # Second part of objective function, using precomputed kernel values

    # element-wise multiplication of alpha_i * y_i
    alpha_y = alpha * y 

    # Matrix multiplication: alpha_y^T @ K gives a row vector; Claude
        # Then @ alpha_y completes the quadratic form
        # This vetorizes the computation rather than needing a nested for loop
    second_term = 0.5 * np.dot(alpha_y, np.dot(K, alpha_y)) 
   
    return second_term - first_term # return negative of sum [ -1(first_term - second_term) ] for minimization
```

```python
# optimized fit function, still accepts kernel arguments, but calculates kernel matrix first then passes that to scipy
def fit_optimized(X, y, kernel_type, C=1.0, gamma=0.1):
    # X represents the matrix of all n samples by m features
    # y represents the vector of target classifications
    # kernel_type denotes what type of kernel operation to perform
    # C represents the regularization parameter
    # gamma represents the RBF kernel hyperparameter
    
    n = X.shape[0] # sample size
    
    # Initialize alphas
    alpha = np.zeros(n)
    
    # Precompute kernel matrix ONCE, this is the main change in logic
    K = compute_kernel_matrix(X, kernel_type, gamma) 
    
    # Establish constraints
    sum_constraint = {'type': 'eq',
                      'fun': lambda alpha: np.dot(alpha, y)
    }
    
    # Using Scipy.optimize.minimize() to find most optimal set of alpha-star's
    result = minimize(
        fun = objective_function_optimized, 
        x0 = alpha,
        args = (K, y),  # Pass K instead of X, kernel_type, gamma
        constraints = [sum_constraint],
        bounds = [(0, C) for _ in range(n)],
        method='SLSQP'
    )

    # retrieve vector of alpha-star's
    alpha = result.x
    

    # Retrieve indices of all support vectors
    indices = np.where(alpha > 1e-5)[0]

    # Call calculate_bias helper function for bias
    bias = calculate_bias(alpha, K, y, indices, C)

    # Store model as a dictionary
    model = {
        'alpha': alpha[indices],
        'sv': X[indices],
        'sv_labels': y[indices],
        'bias': bias,
        'kernel_type': kernel_type,
        'C': C,
        'gamma': gamma
    }
    
    return model # Return our dictionary model that contains all the support vectors and biases   
```

#### Computing Bias

```python
# calculate bias helper function using kernel matrix
def calculate_bias(alpha, K, y, indices, C=1.0):
    # alpha denotes a vector of the alpha_i's or the lagrangian multipliers
    # K represents the precomputed kernel matrix
    # y represents the vector of target classifications
    # indices represents a vector of all the indexes corresponding to a support vector
    # C represents the regularization parameter

    bias_list = [] # Store list of biases
    margin_list = [] # Store list of margin SVs

    # Loop through SVs and obtain all indices of margin support vectors
    for index in indices: 
        if alpha[index] > 1e-5 and alpha[index] < C:
            margin_list.append(index)

    # Catch if no margin support vectors; Claude suggestion
    if len(margin_list) == 0:
        margin_list = indices
    
    # # Loop through and calculate Bias = y_i - Sum(a_j * y_j * K[i,j])
    # for index in margin_list:
        
    #     # Calculate sum_term
    #     sum_term = 0.0
    #     for j in indices:  
    #         sum_term += alpha[j] * y[j] * K[index,j]
        
    #     # Calculate b = y - sum_term
    #     b_i = y[index] - sum_term
    #     bias_list.append(b_i)


    
    # Compute bias for all margin SVs at once using vectorization; Claude

    # Extract relevant rows from kernel matrix for margin support vectors
    K_margin = K[margin_list][:, indices]  # Shape: (n_margin_svs, n_svs)
    
    # Compute sum_j(alpha_j * y_j * K[i,j]) for all margin SVs simultaneously
        # alpha[indices] * y[indices] gives vector of (alpha_j * y_j) for all SVs
        # K_margin @ (alpha[indices] * y[indices]) computes the sum for each margin SV
    sum_terms = np.dot(K_margin, alpha[indices] * y[indices]) 
    
    # Compute b_i = y_i - sum_term for all margin SVs
    bias_values = y[margin_list] - sum_terms

    return np.mean(bias_values)  # Return average of SV biases 
```

#### Prediction

```python
# We make predictions based on the data points from a test set using vectorized operations; Claude assistance
def predict(model, X_new):
    # model is a dictionary containing all of information of our fitted SVM
         # alpha[indices] are all the alpha_i's of support vectors
         # sv is X[indices] are a matrix consisting of all support vector x_i's
         # sv_labels is y[indices], the labels of our SVs
         # bias is the model bias
         # kernel_type is the kernel 
         # C is the regularization parameter
         # gamma is the hyperparameter for RBF Kernel
    # X_new is the data matrix we want to predict on (n_samples, n_features)

    alphas = model['alpha']  # Alpha values for support vectors
    sv = model['sv']  # The actual support vector points (n_sv, n_features)
    sv_labels = model['sv_labels']  # Labels of support vectors
    bias = model['bias']
    kernel_type = model['kernel_type']
    gamma = model['gamma']
    
    # Compute kernel matrix between test points and support vectors
    if kernel_type == 'linear':
        # For linear kernel: K[i,j] = X_new[i]  sv[j]
        # Vectorized as X_new @ sv.T
        K_test = np.dot(X_new, sv.T)
        
    elif kernel_type == 'rbf':
        # For RBF kernel: K[i,j] = exp(-gamma * ||X_new[i] - sv[j]||^2)
        # Using the expansion: ||a - b||^2 = ||a||^2 + ||b||^2 - 2*ab
        
        # Compute squared norms for test points: ||X_new[i]||^2
        sq_norms_new = np.sum(X_new ** 2, axis=1).reshape(-1, 1) 
        
        # Compute squared norms for support vectors: ||sv[j]||^2
        sq_norms_sv = np.sum(sv ** 2, axis=1).reshape(1, -1)
        
        # Broadcasting computes all pairwise squared distances at once
        sq_dists = sq_norms_new + sq_norms_sv - 2 * np.dot(X_new, sv.T)
        
        # Apply RBF transformation
        K_test = np.exp(-gamma * sq_dists) 
    
    else:
        raise ValueError("Current kernel type not supported!")
    
    # Compute decision function for all test points at once
        # decision[i] = bias + sum_j(alpha_j * y_j * K[i,j])
        # This is equivalent to: bias + K_test @ (alphas * sv_labels)
        # where @ is matrix multiplication and * is element-wise multiplication
    
    # Element-wise multiplication of alphas with labels
    alpha_y = alphas * sv_labels 
    
    # Matrix-vector multiplication computes sum for all test points simultaneously
    decisions = bias + np.dot(K_test, alpha_y)
    
    # Convert decision values to class predictions {-1, +1}
    predictions = np.where(decisions >= 0, 1, -1)
    
    return predictions  # Return numpy array of predictions
```

```python
# Train/test split function (70/30 for SVM), adjusted from previous tasks
def train_test_split_svm(X, y, train_ratio=0.70, random_seed=None):
    # X represents the feature matrix size number of samples by number of features
    # y represents the target vector with length equal to number of samples
    # train_ratio represents the proportion of data allocated to training set
    # random_seed is an optional parameter for reproducibility

   
    # Set random seed if provided for reproducibility
    if random_seed is not None:
        np.random.seed(random_seed)
    
    n = X.shape[0]  # Total number of samples
    
    # Create array of indices from 0 to n-1 and shuffle them randomly
    indices = np.arange(n)
    np.random.shuffle(indices)  # In-place randomization

    # Calculate the split points based on ratios
        # Remaining samples go to test set
    train_end = int(n * train_ratio)
    
    # Split the indices into two groups
    train_indices = indices[:train_end]
    test_indices = indices[train_end:]

    # Apply the shuffled indices to X and y to create the splits
    X_train = X[train_indices]
    X_test = X[test_indices]
    y_train = y[train_indices]
    y_test = y[test_indices]
    
    return X_train, X_test, y_train, y_test   # Outputs the split datasets according to shuffled indices where X and y rows align
```

```python
# Function to normalize features, slightly modified from version in Task 2
def normalize_features_svm(X_train, X_test):
    # X_train represents training features
    # X_test represents test features
    
    n_train = X_train.shape[0]
    n_features = X_train.shape[1]
    
    # Calculate means from training data only
    means = np.zeros(n_features)
    for j in range(n_features):
        feature_sum = 0.0
        for i in range(n_train):
            feature_sum += X_train[i, j]
        means[j] = feature_sum / n_train
    
    # Calculate standard deviations from training data only
    stds = np.zeros(n_features)
    for j in range(n_features):
        squared_diff_sum = 0.0
        for i in range(n_train):
            diff = X_train[i, j] - means[j]
            squared_diff_sum += diff * diff
        variance = squared_diff_sum / n_train
        stds[j] = np.sqrt(variance)
    
    # Normalize training set
    X_train_norm = np.zeros_like(X_train)
    for j in range(n_features):
        for i in range(n_train):
            if stds[j] > 1e-8:
                X_train_norm[i, j] = (X_train[i, j] - means[j]) / stds[j]
            else:
                X_train_norm[i, j] = 0.0
    
    # Normalize test set using training statistics to avoid data leakage
    n_test = X_test.shape[0]
    X_test_norm = np.zeros_like(X_test)
    for j in range(n_features):
        for i in range(n_test):
            if stds[j] > 1e-8:
                X_test_norm[i, j] = (X_test[i, j] - means[j]) / stds[j]
            else:
                X_test_norm[i, j] = 0.0
    
    return X_train_norm, X_test_norm   # Outputs both normalized datasets
```

#### Evaluating Performance 

We'll now evaluate our best-performing SVM model using comprehensive metrics: accuracy, precision, recall, F1-score, and ROC-AUC. We'll also create a confusion matrix to visualize the types of errors our model makes. These are the same evaluation metrics we used in Task 2, allowing for direct comparison between our SVM and logistic regression implementations.

We need to convert our SVM predictions from {-1, +1} format to {0, 1} format to use the evaluation functions from Task 2. We'll evaluate all models. 

One thing of note is that we can't pass {-1, +1} as we need to give the ROC-AUC continuous values for it to generate the curve, so we will create a helper function that calculates decision values (which are continuous values) for each point in order to utilize the compute_roc_auc function from Task 1.

```python
# Function to compute decision scores for ROC curve
def predict_scores(model, X_new):
    # model is a dictionary containing all of information of our fitted SVM
         # alpha[indices] are all the alpha_i's of support vectors
         # sv is X[indices] are a matrix consisting of all support vector x_i's
         # sv_labels is y[indices], the labels of our SVs
         # bias is the model bias
         # kernel_type is the kernel 
         # C is the regularization paramter
         # gamma is the hyperparamter for RBF Kernel
    # X_new represents new data points to score

    # Extract values from model dictionary
    alphas = model['alpha']
    sv = model['sv']
    sv_labels = model['sv_labels']
    bias = model['bias']
    kernel_type = model['kernel_type']
    gamma = model['gamma']
    
    scores = []  # empty array to store decision scores for each point
    
    # loop through every point in testing data
    for current_pt in X_new:
        
        decision_sum = bias  # Track f(current_pt)
        
        # loop to sum a_i * y_i * K(x_i,x_new)
        for i in range(len(sv)):
            decision_sum += alphas[i] * sv_labels[i] * kernel(sv[i], current_pt, kernel_type, gamma)
        
        # Store decision score to vector
        scores.append(decision_sum)
    
    return np.array(scores)  # return vector of decision scores
```

```python
# Function to evaluate SVM model performance using Model Evaluation functions from Task 2
    # We do need to change both the predicted and actual y's to {0,1} from {-1,1} to reuse  
def evaluate_svm_model(model, X_test, y_test):
    # model is the trained SVM model dictionary
    # X_test represents the test features
    # y_test represents the test labels in {-1, +1} format
    
    # Make predictions on test set
    y_test_pred = predict(model, X_test)
    
    # Get decision scores for ROC-AUC calculation
    y_test_scores = predict_scores(model, X_test)
    
    # Convert SVM predictions from {-1, +1} to {0, 1} in order to use Task 2's metric functions
    final_y_pred = np.zeros(len(y_test_pred))
    for i in range(len(y_test_pred)):
        if y_test_pred[i] == 1:
            final_y_pred[i] = 1
        else:
            final_y_pred[i] = 0
    
    final_y_pred = final_y_pred.astype(int) # Cast back to int; Claude
    
    # Convert test labels from {-1, +1} to {0, 1}
    y_test_binary = np.zeros(len(y_test))
    for i in range(len(y_test)):
        if y_test[i] == 1:
            y_test_binary[i] = 1
        else:
            y_test_binary[i] = 0
    
    y_test_binary = y_test_binary.astype(int) # Cast back to int; Claude
    
    # Calculate all evaluation metrics
    accuracy = compute_accuracy(y_test_binary, final_y_pred)
    precision = compute_precision(y_test_binary, final_y_pred)
    recall = compute_recall(y_test_binary, final_y_pred)
    f1 = compute_f1_score(y_test_binary, final_y_pred)
    roc_auc = compute_roc_auc(y_test_binary, y_test_scores)
    cm = compute_confusion_matrix(y_test_binary, final_y_pred)

    
    # Print results
    print("Test Set Evaluation Results:")
    print("Accuracy: ", accuracy)
    print("Precision: ", precision)
    print("Recall: ", recall)
    print("F1-Score: ", f1)
    print("ROC-AUC: ", roc_auc)
    print("Confusion Matrix: \n", cm)
```

### Analysis

So now that we've come up with our chosen model for predicting breast cancer, we will go into further analysis into how the hyperparameters affected our SVM models, visualize our support vectors and decision boundaries, as well as compare how our RBF SVM model compares to Scikit-Learn's SVM and our previous Logistic Regression model. 

#### How different hyperparameters affect decision boundaries

The regularization parameter C controls the trade-off between achieving a wide margin and correctly classifying training points. Geometrically, C determines how much we penalize misclassifications during training.

When C is small (like C=0.1), we prioritize having a wide margin between classes, even if that means some training points fall on the wrong side of the margin or are misclassified. This creates a "soft margin" that's more tolerant of outliers and noise in the training data. The decision boundary tends to be smoother and more generalizable because the model doesn't contort itself to fit every single training point perfectly.

When C is large (like C=10.0), we heavily penalize any misclassifications, forcing the model to work harder to correctly classify every training point. This creates a "hard margin" where the decision boundary can become more complex and irregular as it bends around individual points. While this achieves higher training accuracy, it risks overfitting to noise in the training data.

For our linear kernel results, we saw this firsthand as C=0.1 achieved the best test performance (F1=97.64%), while C=1.0 and C=10.0 performed slightly worse. However, for the purposes of balancing concerns of overfitting and underfitting, we chose C=1.0 as a "balanced" hyperparameter value when fitting for our non-linear RBF SVM. 
<br>
<center>*************************************</center>
<br>
The gamma parameter in RBF kernels controls the "reach" or "influence" of each support vector. Geometrically, each support vector creates a Gaussian "bump" in the feature space, and gamma determines the width of these bumps.

When gamma is small (like gamma=0.01), each support vector influences points far away from it. The Gaussian bumps are wide and overlap significantly, creating smooth, gradual transitions in the decision boundary. This is similar to having a "global" view where the model considers broader patterns in the data.

When gamma is large (like gamma=1.0), each support vector only influences points very close to it. The Gaussian bumps are narrow and isolated, creating sharp, localized peaks in the feature space. This is similar to having a "local" view where the model focuses intensely on individual points.

Mathematically, recall that the RBF kernel is: $K(x_i, x_j) = \exp(-\gamma ||x_i - x_j||^2)$

When $\gamma$ is large, the exponential term becomes very small (close to zero). This means points need to be extremely close to a support vector to be influenced by it. The decision boundary becomes a collection of tiny "islands" around each support vector rather than a smooth, continuous surface.

We see this play out as we increase the value of Gamma: 

- Gamma=0.01: Wide bumps create smooth boundaries and good generalization (F1=97.64%, Recall=98.41%)
- Gamma=0.1: Medium bumps create more complex boundaries, which slightly degraded performance (F1=93.23%, Recall=98.41%)  
- Gamma=1.0: Narrow bumps fragment the space, leading to abysmal performance and catastrophic failure (F1=3.13%, Recall=1.59%)

The transition from gamma=0.1 to gamma=1.0 is particularly dramatic. While gamma=0.1 maintained reasonable recall, gamma=1.0 so severely overfit, it only labeled *one* data point as malignant. The ROC-AUC value remaining high at 95.41% seems contradictory, but it actually isn't. The decision scores still somewhat correlate with the true labels, but the classification threshold was so bad that no data point successfully crosses it. 

#### Visualize support vectors and decision boundaries

To visualize our SVM decision boundaries and support vectors, we will reduce our 30-dimensional feature space down to 2 dimensions. We accomplish this using Principal Component Analysis (PCA), which identifies the two directions in our data that capture the most variance. By projecting our data onto these two principal components, we can create 2D visualizations while preserving as much information as possible about the structure and separability of our classes.

```python
# Function to perform PCA using SVD method; Claude assistance
def compute_pca(data, n_components=2):
    # data represents the input matrix
    # n_components is the number of principal components to keep
    
    
    # Calculate mean for each feature
    col_mean = np.mean(data, axis=0)
    
    # Calculate standard deviation for each feature
    col_std = np.std(data, axis=0)
    
    # Standardize: (X - mu) / sigma
    standardized_data = (data - col_mean) / col_std

    
    # SVD decomposes X = U @ S @ V.T
    # V contains our principal components (no covariance matrix needed!)
    U, singular_values, VT = np.linalg.svd(standardized_data, full_matrices=False)


    # VT has principal components as rows, we want them as columns
    principal_components = VT.T
    
    # Select top n_components
    components_subset = principal_components[:, 0:n_components]
    
    # Project data onto principal components
    projected_data = np.dot(standardized_data, components_subset)

    # Eigenvalues = (singular_values^2) / (n_samples - 1)
    n_samples = data.shape[0]
    eigenvalues = (singular_values ** 2) / (n_samples - 1)
    
    # Calculate explained variance
    total_variance = np.sum(eigenvalues)
    explained_variance = eigenvalues / total_variance
    cumulative_variance = np.cumsum(explained_variance)

    
    return projected_data, components_subset, col_mean, col_std
```

```python
# Apply PCA to reduce training data to 2 dimensions
X_train_2d, components, train_mean, train_std = compute_pca(X_train_norm, n_components=2)

# Transform test data using the same principal components from training
# This ensures consistency and avoids data leakage
X_test_standardized = (X_test_norm - train_mean) / train_std
X_test_2d = np.dot(X_test_standardized, components)
```

Note that the performance on this reduced data will be lower than our full 30-dimensional models since we're only capturing a portion of the total variance. However, this 2D representation allows us to visualize the decision boundaries, support vectors, and how different hyperparameters affect the geometry of the classification boundary.

```python
# Function to visualize SVM decision boundary and support vectors in 2D; Heavy Claude Assistance
def plot_svm_decision_boundary(model, X_train, y_train, title):
    # model is the trained SVM model dictionary
    # X_train is the 2D training data
    # y_train is the training labels in {-1, +1} format
    # title is the plot title string
    
    # Determine the bounds of our plot based on training data
    x_min = X_train[:, 0].min() - 1
    x_max = X_train[:, 0].max() + 1
    y_min = X_train[:, 1].min() - 1
    y_max = X_train[:, 1].max() + 1
    
    # Create a mesh grid with fine spacing
    h = 0.02  # Step size in the mesh
    xx_vals = np.arange(x_min, x_max, h)
    yy_vals = np.arange(y_min, y_max, h)
    
    # Create meshgrid from x and y values
    xx, yy = np.meshgrid(xx_vals, yy_vals)
    
    # Flatten the grid to pass through predict function
    grid_points = np.c_[xx.ravel(), yy.ravel()]
    
    # Get predictions for all grid points
    Z = predict(model, grid_points)
    
    # Convert predictions list to numpy array
    Z = np.array(Z)
    
    # Reshape predictions back to grid shape for plotting
    Z = Z.reshape(xx.shape)
    
    # Create the plot
    plt.figure(figsize=(10, 8))
    
    # Plot decision regions with color fill
    # Benign region (class -1) in light blue, Malignant region (class +1) in light coral
    plt.contourf(xx, yy, Z, alpha=0.3, levels=[-1, 0, 1], colors=['lightblue', 'lightcoral'])
    
    # Plot decision boundary as a thick black line where Z=0
    plt.contour(xx, yy, Z, levels=[0], colors='black', linewidths=2)
    
    # Plot training points
    for i in range(len(X_train)):
        if y_train[i] == 1:
            # Malignant samples as red circles
            plt.scatter(X_train[i, 0], X_train[i, 1], c='red', marker='o', s=50, edgecolors='k', linewidths=1)
        else:
            # Benign samples as blue squares
            plt.scatter(X_train[i, 0], X_train[i, 1], c='blue', marker='s', s=50, edgecolors='k', linewidths=1)
    
    # Highlight support vectors with larger yellow circles
    sv = model['sv']
    sv_labels = model['sv_labels']
    
    for i in range(len(sv)):
        plt.scatter(sv[i, 0], sv[i, 1], c='none', s=200, 
                   edgecolors='yellow', linewidths=3, alpha=0.8)
    
    plt.xlabel('First Principal Component', fontsize=12)
    plt.ylabel('Second Principal Component', fontsize=12)
    plt.title(title, fontsize=14)
    plt.grid(True, alpha=0.3)
    
    # Create legend
    from matplotlib.patches import Patch
    legend_elements = [
        Patch(facecolor='lightcoral', edgecolor='black', label='Malignant Region'),
        Patch(facecolor='lightblue', edgecolor='black', label='Benign Region'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', 
                   markersize=10, markeredgecolor='k', label='Malignant Training'),
        plt.Line2D([0], [0], marker='s', color='w', markerfacecolor='blue', 
                   markersize=10, markeredgecolor='k', label='Benign Training'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='none', 
                   markersize=15, markeredgecolor='yellow', markeredgewidth=3, label='Support Vector')
    ]
    plt.legend(handles=legend_elements, loc='best', fontsize=10)
    
    plt.tight_layout()
    plt.show()
```

```python
# Visualize Linear SVM with C=0.1
plot_svm_decision_boundary(model_2d_linear_c01, X_train_2d, y_train, 
                           'Linear SVM with C=0.1')
```
